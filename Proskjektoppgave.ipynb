*Det er laget funksjoner som skal vises i figuren som; Likevektspris og likevektesmengde, konsumentoverskudd, produsentoverskudd, samfunnsøkonomisk overskudd og de linære kurvene for tilbud og etterspørsel.*
*Konsumentoverskudd er betalingsvilligheten minus det man faktisk må betale og produsentoverskudd er differansen mellom inntekt og variable kostander. Det samfunnsøkonomiske overskuddet er PO og KO til sammen, og er en måte å vurdere om hva som er samfunnsøkonomisk optimalt vis det samfunnsøkonomiske overskuddet er maksimalt. Hvis det ikke er optimalt pga. endringer fra konsument eller produsent vil det oppstå et samfunnsøkonomisk tap.*

```python
class model:
    
    def __init__ (self, a, b, d, s):
        self.a, self.b, self.d, self.s = a, b, d, s
        if a < b:
            raise ValueError("insufficient demand.")

        
    def eq(self):
        return (self.a - self.b) / (self.d + self.s) #Equilibrium quantity

    def ep(self):
        return (self.a - self.d * self.eq()) #Equilibrium price

    def c_surplus(self):
        return (self.a - self.ep())*self.eq() /2 #Consumer surplus

    def p_surplus(self):
        return (self.ep() - self.b) * self.eq() /2 #Producer surplus

    def totalsurplus(self):
        return self.p_surplus() + self.c_surplus() #Total surplus

    def demand(self, x):
        return self.a - self.d*x #Demand-curve

    def supply(self, x):
        return self.b + self.s*x #Supply-curve    
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


#baseline: a,b, d, s
baseline_numbers = 10, 0, 1, 1
m = model (*baseline_numbers)

q_max = m.eq() * 2
q_grid = np.linspace(0.0, q_max, 100)
pd = m.demand(q_grid)
ps = m.supply(q_grid)

fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(q_grid, pd, lw=3, alpha=0.9, label="Etterspørsel", color="b")
ax.plot(q_grid, ps, lw=3, alpha=0.9, label="Tilbud", color="r")
ax.set_xlabel("Kvantum", fontsize=15)
ax.set_xlim(0, q_max)
ax.set_ylabel("Pris", fontsize=15)
ax.legend(loc="lower right", frameon=False, fontsize=12)
ax.set(title="Tilbud og etterspørsel i et marked")
plt.show()
```

**En enkel tilbuds-etterspørsels modell i likevekt**

> Utregningen av likevektspris-, og kvantum, KO, PO, SO gjøres slik:

```python
# Baseline numbers,a ,b, d, s
baseline_numbers = 10,0,1,1
m = model(*baseline_numbers)

print("Produsentoverskudd er", m.p_surplus())
print("Konsumentoverskudd er", m.c_surplus())
print("Samfunnsøkonomisk overskudd er ", m.totalsurplus())
print("Likevektskvantum er ", m.eq())
print("Likevektspris er ", m.ep())
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(q_grid, pd, lw=3, alpha=0.9, label="Etterspørsel", color="b")
ax.plot(q_grid, ps, lw=3, alpha=0.9, label="Tilbud", color="r")
ax.set_xlabel("Kvantum", fontsize=15)
ax.set_xlim(0, q_max)
ax.set_ylabel("Pris", fontsize=15)
ax.legend(loc="lower right", frameon=False, fontsize=12)
ax.set(title="Tilbud og etterspørsel i et marked")
ax.axhline(5,0,0.5,linewidth=2, color="black", linestyle = "dashed")
ax.axvline(5,0,0.5,linewidth=2, color="black",linestyle="dashed")
ax.text(5.3, 5, "Likevektspunktet",fontsize=12)
ax.text(0.6, 6, "Konsumoverskudd",fontsize=12)
ax.text(0.6, 3, "Produsentoverskudd",fontsize=12)
ax.text(0.2, 5.2, "$P^{*}$",fontsize=14)
ax.text(5.1, 0, "$Q^{*}$",fontsize=14)
```

*En modell i likevekt der tilbud = etterspørsel vil optimal pris (P*) være 5 og kvantum (Q*) *være 5. Produsentoverskuddet og kosnumentoverskuddet vil være like store og det samfunnsøkonomiske overskuddet vil være optimalt.*

*Markedet sørger selv for å komme i likevekt ved at:*

*1)Konsumentene kjøper ikke et produkt hvis hvis dem synes at prisen er for høy.*

*2) Konsumentene velger å kjøpe et billigere alternativ.*

*3) Når det er udekket etterspørsel, driver dette prisene opp.*

*4) Når prisen øker blir det mer attraktivt å produsere varen. Økt pris gjør at også mindre effektive selskaper også kan få en fortjeneste på produksjonen. De allerede effektive selskapene vil naturlig nok øke produksjonen (hvis det er nok kapasitet), for da øker de inntjeningen.*

### Skift i modellen: <a class="anchor" id="skiftimodell"></a>



**En bedrift setter opp prisen på produktet sitt, og vil ikke selge et produkt for mindre enn 5. Det gjør at etterspørselen etter produktet går ned, og det blir et skift i likevekten mellom pris og etterspørsel til venstre i modellen**

```python
#baseline numbers: a,b, d, s
baseline_numbers = 10, 5, 1, 1
m = model (*baseline_numbers)

q_max = m.eq() * 3
q_grid = np.linspace(0.0, q_max, 100)
pd = m.demand(q_grid)
ps = m.supply(q_grid)

fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(q_grid, pd, lw=3, alpha=0.9, label="Etterspørsel", color="b")
ax.plot(q_grid, ps, lw=3, alpha=0.9, label="Tilbud", color="r")
ax.set_xlabel("Kvantum", fontsize=15)
ax.set_xlim(0, q_max)
ax.set_ylabel("Pris", fontsize=15)
ax.legend(loc="lower right", frameon=False, fontsize=12)
ax.set(title="Tilbud og etterspørsel i et marked")
ax.axhline(7.5,0,0.33,linewidth=2, color="black", linestyle = "dashed")
ax.axvline(2.5,0,0.5,linewidth=2, color="black",linestyle="dashed")
ax.text(2.7, 7.4, "Likevektspunktet",fontsize=12)
ax.text(0.05, 8.2, "Konsumoverskudd",fontsize=12)
ax.text(0.05, 7, "Produsentoverskudd",fontsize=12)
ax.text(0.0, 7.6, "$P^{*}$",fontsize=14)
ax.text(2.7, 2.5, "$Q^{*}$",fontsize=14)



plt.show()
```

> Her ser man at Likevektspris-, og kvantum tilpasser seg den nye likevekten, KO og PO blir mindre, prisen øker og kvantum går ned, noe som resulterer i et SO-tap, siden SO ikke er optimal.

### IS-kurven: <a class="anchor" id="iskurven"></a>

*IS-Kurven bygger på Keynesmodellen i en åpen økonomi. Keynes modellen for en åpen økonomi blir brukt til beregninger av  størrelsen på BNP. Konsekvensanalyse i hvordan likevekten blir endret dersom eksogene variabler eller parametere blir endret. I modellen er etterspørselssiden reprensetert ved en IS-ligning. Realligningen i en åpen økonmomi er Y = C + I + G + NX, som betyr at BNP Y er lik summen av konsum(C), private investeringer(I), offentlig bruk av varer og tjenester (G) og nettoeksporten(NX = X-Q).* 

*Sentrale forutsetninger i modellen er gitt ved:* 


*-Kort sikt* 

*-Etterspørselsbestemt produksjon*  

*-Priser og lønninger er gitt* 

*-Statisk model.*

*IS-RR-PK-kruven i en åpen økonom som i denne oppgaven har flytende valutakurs, og inflasjonsmål. Sentralbanken vil holde infalsjon nær inflasjonsmålet, og stabilisere BNP nær sitt potensielle nivå. Handlingsmønsteret til sentral banken vil være å bruke renteregelen for å justere mot inflasjonsmålet = 2%*
(inflasjonmålet var fra 2001 frem til 2018 på 2,5%, før det ble satt til 2%) [7](https://snl.no/inflasjonsm%C3%A5l)
*I modellen nedenfor vil det være egendefinerte matematiskeregnestykker med hold i de orginale ligningene, siden IS-RR vil orginalt ikke gi oss en linær linje så det vil være en forenklet modell.*

> *I modellen definerer vi variabler og funksjoner*

```python
import numpy as np               
import matplotlib.pyplot as plt 

#Define parameters:
# Parameters
Y = 100
z = 40                 # Autonomous consumption
c = 0.2                # Marginal propensity 
Q = 5                  # Import
b  = 0.2               # Marginal propensity 
T = 1                  # Taxes
I = 10                 # Investment 
G = 8                  # Government spending
X = 2                  # Exports (given)
d = 10                 # Investment parameter

Y = np.arange(Y)  

# Define and popultat the IS-Curv
def i_IS(z, Q, c, b, T, I, G, X, d, Y):
    i_IS = ((z-Q)-(c-b)*T + I + G + X - (1-c+b)*Y)/d
    return i_IS

def Y_IS(z, Q, c, b, T, I, G, X, d, i):
    Y_IS = ((z-Q)-(c-b)*T + I + G + X - d*i)/(1-c+b)
    return Y_IS

i = i_IS(z, Q, c, b, T, I, G, X, d, Y)


y_max = np.max(i)
x_max = Y_IS(z, Q, c, b, T, I, G, X, d, 0)

fig, ax = plt.subplots(figsize=(8, 6))
ax.set(title="IS-Kurve", xlabel=r'BNP(Y)', ylabel=r'Rente(r)')
ax.plot(Y, i, "b")
plt.xticks([])

plt.show()
```

**Teorien bak IS-kurven er jo lavere renten blir, jo større vil BNP bli, og jo høyere renten blir jo lavere vil BNP bli.**

### RR-kurven: <a class="anchor" id="rrkurven"></a>


*RR-kurven viser sentralbankens rentesetting ved et inflasjonsmål, som vi antar følger et fast handlingsmønster. Norges Bank bruker konsumprisindeksen (KPI) til å måle inflasjonen, og KPI har noen priser som har en tedens til å variere mye fra en periode til en annen, som f.eks. energipriser kan stige mye i en periode og falle i en annen periode. Slike prisendringer kan skape støy i den underliggende trenden i prisutviklingen og derfor bruker Norges bank ulike indikatorer for å se gjennom forbigående svigninger i inflasjonen. Den viktigste indikatoren for underliggende inflasjon i analysene er KPI justert for avgiftsendring og uten energivarer(KPI - JAE). Målet for pengepolitikken er at den årlige veksten i konsumprisene/inflasjonen er nær 2 prosent over tid.*

*Handlingsmønstert til sentralbanken ved renteregelen innbærer at de vil sette en høy rente hvis inflasjonen er høyere enn inflasjonsmålet (likevekten i IS-RR-modell), eller hvis BNP er større enn potensielt BNP. Renten vil bidra til at infalsjonen reduseres ned mot målet, i tillegg vil en renteheving føre til at kornekursen styrkes, slik at prisveksten på importvarer dempes, som igjen fører til lavere vekst i konsumprisene.*




```python
import numpy as np               
import matplotlib.pyplot as plt  

# Define parameters
# Parameters:
Y = 100
c1 = 0.4             # Parameter
c2 = 1               # Parameter
c3 = 8               # Parameter
Ms = 200             #Nominal money 
P  = 15              #Price level

Y = np.arange(Y)  

#Define and populate the RR-Curve
def i_RR(c1, c2, c3, Ms, P, Y):
    i_RR = (c1 - Ms/P)/c3 + c2/c3*Y
    return i_RR

i = i_RR(c1, c2, c3, Ms, P, Y)
    
#Plot the RR-curve
y_max = np.max(i)
                      
fig, ax = plt.subplots(figsize=(8, 6))
ax.set(title="RR-kurve", xlabel=r'BNP(Y)', ylabel=r'rente(r)')
ax.plot(Y, i, "r")  
plt.xticks([])
plt.show()
```

**Funksjonen bak RR-kurven er at den følger IS-kruven, for å styre rentenivået etter hvilket eksogen handling det skjer på parameterene**

### Modellene slått sammen: <a class="anchor" id="modelleneslåttsammen"></a>



```python
import numpy as np               
import matplotlib.pyplot as plt  

#Defin parameters:
Y = 100
Y = np.arange(Y)     

# IS Parameters
z = 40                    # Autonomous consumption
c = 0.2                   # Marginal propensity to consum 
Q = 5                     # Import
b = 0.2                   # Marginal propensity 
T = 1                     # Taxes
I = 40                    # Investment 
G = 8                     # Government spending
X = 2                     # Exports 
d = 10                    # Investment parameter

# RR Parameters
c1 = 0.4                 # Parameter
c2 = 1                   # Parameter
c3 = 8                   # Parameter                  
Ms = 200                 # Nominal money supply
P  = 15                  # Price level

#Define and populate IS-Curve and RR-Curve:
def i_IS(z, Q, c, b, T, I, G, X, d, Y):
    i_IS = ((z-Q)-(c-b)*T + I + G + X - (1-c+b)*Y)/d
    return i_IS

def i_RR(c1, c2, c3, Ms, P, Y):
    i_RR = (c1 - Ms/P)/c3 + c2/c3*Y
    return i_RR

iIS = i_IS(z, Q, c, b, T, I, G, X, d, Y)
iRR = i_RR(c1, c2, c3, Ms, P, Y)

#Calculate values: 

Y_1 = ((z-Q) - (c-b)*T + I + G + X)/d
Y_2 = (1/c3) * (c1 - Ms/P)
Y_3 = (1 - c + b)/d + (c2/c3)
Y_x  = (Y_1 - Y_2)/Y_3

i_1 = (1/c3)*(c1 - Ms/P)
i_2 = (c2/c3)*Y_x
i_y  = i_1 + i_2


#Plot the IS-RR model:                     
fig, ax = plt.subplots(figsize=(8, 6))
ax.set(title="IS-RR MODEL", xlabel=r'BNP(Y)', ylabel=r'rente(r)')
ax.plot(Y, iIS, "b-")                   #IS-curv
ax.plot(Y, iRR, "r-")                   #RR-curv
plt.plot(Y_x, i_y, 'ko')                # Equilibrium point
plt.axvline(x=Y_x, ls=':', color='k')   # Equilibrium line
plt.axhline(y=i_y, ls=':', color='k')   # Equilibrium line
plt.xticks([])

plt.show()
```

**I modellen ser man likevekten for en IS-RR-kurve**

### PK-kurven: <a class="anchor" id="pkkurven"></a>
​
*PK-kurven også omtalt som Philips-kurven viser fra sin oppringeligeform sammenheng mellom lønnsstigning og arbeidsledighet dvs. at når arbeidsleidgheten synker så stiger inlfasjonen.
PK-kurven skal vise sammenhengen mellom to markoøknomiske variabler som er infalsjon og BNP-gapet(produksjonsgapet). Inflasjonen blir avgjort om hvordan BNP-gapet oppfører seg, øker inflasjonen så vil hele PK-kurven flytte seg et hakk opp som vil gjøre at BNP-gapet synker, og omvendt for å holde BNP i sjakk vis den øker. En PK-kruve som viser sammenhengen mellom lønnstigning og arbeidsledighet vil være omvendt av PK-kurven nedenfor.*

```python
import matplotlib.pyplot as plt
import numpy as np

#Calculate values
t = np.arange(0.01, 5, 0.01)          
s = np.exp(-t)
plt.subplots(figsize=[8, 6])
plt.plot(t, s, "g")

plt.xlim(5, 0) #inflasion prosent
plt.ylim(0, 1)

plt.axvline(x=2.5,ls=":", color = "k")  # Equilibrium line
plt.xlabel('BNP (Y)')
plt.ylabel('Inflasjon (π)')
plt.title('PK_kurven')
plt.xticks([])
plt.plot(figsize=[20, 8])


plt.show()
```

**I PK-kruven kan man se at inflasjonmålet er der likevekten i modellen er. Et skift opp, ned, høyre eller venstre vil øke gapet fra inflasjonsmålet.**
